## pyplot vs matplotlib OO method?
* when plotting something quickly use pyplot
* hen plotting something more advanced use OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
len(over_50.head())
```

```python
# pyplot method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target");
```

```python
# OO method mixed with pyplot
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
            ax=ax);

#ax.set_xlim([45,100])
```

```python
over_50.target.values
```

```python
# OO from scratch
fix, ax = plt.subplots(figsize=(10,6))
# plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);
# customize the plot
ax.set(title="cholestrol level",
       xlabel="age",
       ylabel="chol")
# add a legend
ax.legend(*scatter.legend_elements(),title="target");
# add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
# subplot of chol,age,thalach
fig,(ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                            sharex=True)
# add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);
# customize the plot
ax0.set(title="cholestrol level",
       xlabel="age",
       ylabel="chol")
# add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="target");
# add a meanline
ax0.axhline(y=over_50["chol"].mean(),
          linestyle='--');

# add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"]);
# customize the plot
ax1.set(title="cholestrol level",
       xlabel="age",
       ylabel="heart rate")
# add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="target"); #ax.legend().set_visible(True)
# add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
          linestyle='--');

# add a title to figure
fig.suptitle("heart disease analysis", fontsize=16, fontweight="bold");
```

```python
import matplotlib.pyplot as plt
#Example 1: Basic Pie Chart
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)

#Example 2: Pie Chart with Percentage Labels
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%')

#Example 3: Custom Colors for Pie Chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels,
       colors=['olivedrab', 'rosybrown', 'gray', 'saddlebrown'])

#Example 4: Pie Chart with Hatch Patterns
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, hatch=['**O', 'oO', 'O.O', '.||.'])

#Example 5: Customizing Percentage and Label Distance
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%',
       pctdistance=1.25, labeldistance=.6)

#Example 6: Exploding a Slice in the Pie Chart
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
       shadow=True, startangle=90)
plt.show()

#Example 7: Custom Text Properties and Radius
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%.0f%%',
       textprops={'size': 'smaller'}, radius=0.5)
plt.show()

#Example 8: Shadow and Start Angle Adjustments
fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
       shadow={'ox': -0.04, 'edgecolor': 'none', 'shade': 0.9}, startangle=90)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-5, 5, 100)  # Generate 100 points between -5 and 5

# Define a function to plot (e.g., y = x^2)
y = x**2

# Plot the curve
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x^2', color='blue', linewidth=2)

# Add labels and title
plt.title('Plot of y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-2*np.pi, 2*np.pi, 100)

# Define functions to plot
y1 = np.sin(x)
y2 = np.cos(x)

# Plot the curves
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='y = sin(x)', color='blue', linewidth=2)
plt.plot(x, y2, label='y = cos(x)', color='red', linewidth=2)

# Add labels and title
plt.title('Plot of sin(x) and cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Data
sizes = [15, 30, 45, 10]
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
explode = (0, 0.1, 0, 0)  # explode the 2nd slice (i.e. 'Hogs')

# Plot
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(aspect="equal"))

wedges, texts, autotexts = ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
                                  shadow=True, startangle=140)

for wedge in wedges:
    wedge.set_edgecolor('w')

plt.setp(autotexts, size=10, weight="bold")

ax.set_title("3D-like Pie Chart Example")

plt.show()
```
